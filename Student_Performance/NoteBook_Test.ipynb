Loading to the Data Base:

```python
class ToMongo():


    def __init__(self):

        load_dotenv()

        self.mongo_url = os.getenv('MONGO_URL')

        self.client = pymongo.MongoClient(self.mongo_url)

        self.db = self.client.db

        self.Student_per = self.db.Student_per



    def upload_one_by_one(self):
        
        for i in self.db.index: 

            self.Student_per.insert_one(self.df.loc[i].to_dict()) 



if __name__ == '__main__':

    c = ToMongo()

    print(c.upload_one_by_one)
```
